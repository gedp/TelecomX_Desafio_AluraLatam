```python
# ===================================================================
# DESAFÍO TELECOM X - PROCESO ETL COMPLETO
# ===================================================================

# Importación de bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import requests
from urllib.request import urlopen
import warnings
warnings.filterwarnings('ignore')

# Configuración para visualizaciones
plt.style.use('ggplot')
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = (12, 8)

# ===================================================================
# 1. EXTRACCIÓN DE DATOS (E - EXTRACT)
# ===================================================================

print("="*60)
print("ETAPA 1: EXTRACCIÓN DE DATOS (EXTRACT)")
print("="*60)

# URL de la API de Telecom X (actualizada según el desafío)
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json"

print(f"URL de la API: {url}")

# Cargar datos desde la API
try:
    response = requests.get(url)
    response.raise_for_status()  # Verificar si la solicitud fue exitosa
    data = response.json()
    print("✅ Datos cargados exitosamente desde la API")
except Exception as e:
    print(f"❌ Error al cargar los datos: {e}")
    # Intentar con método alternativo
    try:
        response = urlopen(url)
        data = json.loads(response.read())
        print("✅ Datos cargados exitosamente con método alternativo")
    except Exception as e2:
        print(f"❌ Error con método alternativo: {e2}")
        data = None

# Convertir a DataFrame de pandas
if data:
    df = pd.DataFrame(data)
    print(f"✅ DataFrame creado con {df.shape[0]} filas y {df.shape[1]} columnas")

    # Visualizar las primeras filas
    print("\nPrimeras 5 filas del dataset:")
    display(df.head())

    # Información general del dataset
    print("\nInformación general del dataset:")
    df.info()

    # Estadísticas básicas
    print("\nEstadísticas básicas:")
    display(df.describe(include='all'))
else:
    print("❌ No se pudieron cargar los datos. Verifica la URL o la conexión a internet.")

# ===================================================================
# 2. TRANSFORMACIÓN DE DATOS (T - TRANSFORM)
# ===================================================================

if data:
    print("\n" + "="*60)
    print("ETAPA 2: TRANSFORMACIÓN DE DATOS (TRANSFORM)")
    print("="*60)

    # 2.1 Exploración inicial y comprensión de los datos
    print("\n2.1 Exploración inicial del dataset:")

    # Flattening nested dictionary columns
    print("\nFlattening nested dictionary columns:")
    nested_cols = ['customer', 'phone', 'internet', 'account', 'Charges'] # Added 'Charges'
    for col in nested_cols:
        if col in df.columns:
            # Flatten the column and concatenate, then drop the original
            df = pd.concat([df.drop(col, axis=1), df[col].apply(pd.Series)], axis=1)
            print(f"✅ Columna '{col}' aplanada.")
        else:
            print(f"⚠️ Columna '{col}' no encontrada para aplanar.")


    # Verificar valores únicos por columna
    print("\nValores únicos por columna:")
    for col in df.columns:
        print(f"{col}: {df[col].nunique()} valores únicos")
        if df[col].nunique() < 10:
            print(f"  Valores: {df[col].unique()}")

    # 2.2 Limpieza de datos
    print("\n2.2 Limpieza de datos:")

    # Verificar valores nulos
    print("\nVerificación de valores nulos:")
    null_values = df.isnull().sum()
    null_percentage = (null_values / len(df)) * 100
    null_info = pd.DataFrame({
        'Valores Nulos': null_values,
        'Porcentaje': null_percentage
    })
    display(null_info[null_info['Valores Nulos'] > 0])

    # Verificar filas duplicadas
    duplicate_rows = df.duplicated().sum()
    print(f"\nFilas duplicadas: {duplicate_rows}")

    # Tratamiento de valores nulos
    # Estrategia según el tipo de columna
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype == 'object':
                # For categorical columns, fill with mode
                mode_value = df[col].mode()[0]
                df[col].fillna(mode_value, inplace=True)
                print(f"✅ Columna '{col}': Valores nulos rellenados con la moda: '{mode_value}'")
            else:
                # For numerical columns, fill with median
                median_value = df[col].median()
                df[col].fillna(median_value, inplace=True)
                print(f"✅ Columna '{col}': Valores nulos rellenados con la mediana: {median_value}")

    # Eliminar filas duplicadas si existen
    if duplicate_rows > 0:
        df = df.drop_duplicates()
        print(f"✅ Filas duplicadas eliminadas. Nuevo tamaño: {df.shape}")

    # 2.3 Transformación de datos
    print("\n2.3 Transformación de datos:")

    # Identify categorical and numerical columns
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    print(f"\nColumnas categóricas: {categorical_cols}")
    print(f"Columnas numéricas: {numeric_cols}")

    # Crear columna de cuentas diarias (si existe una columna de facturación mensual)
    billing_keywords = ['mensual', 'monthly', 'cargo', 'charge', 'pago', 'payment', 'factura', 'bill', 'Monthly']
    billing_cols = []
    for col in df.columns:
        if any(keyword in col.lower() for keyword in billing_keywords):
            billing_cols.append(col)

    if 'Monthly' in df.columns: # Prefer 'Monthly' after flattening 'Charges'
         df['Cuentas_Diarias'] = pd.to_numeric(df['Monthly'], errors='coerce') / 30
         print(f"✅ Columna 'Cuentas_Diarias' creada a partir de 'Monthly'")
    elif billing_cols:
        billing_col = billing_cols[0]  # Use the first billing column found
        df['Cuentas_Diarias'] = pd.to_numeric(df[billing_col], errors='coerce') / 30
        print(f"✅ Columna 'Cuentas_Diarias' creada a partir de '{billing_col}'")
    else:
        print("⚠️ No se encontró una columna de facturación para crear 'Cuentas_Diarias'")

    # Convert text values to binary (e.g., "Yes"/"No" to 1/0)
    binary_mappings = {
        'Sí': 1, 'Si': 1, 'Yes': 1, 'YES': 1, 'yes': 1, 'True': 1, 'true': 1,
        'No': 0, 'NO': 0, 'no': 0, 'False': 0, 'false': 0, '': np.nan
    }

    for col in categorical_cols:
        if col in df.columns: # Check if column still exists after flattening/dropping
            unique_vals = df[col].unique()
            # Check if the column has binary values
            if any(val in binary_mappings.keys() for val in unique_vals):
                # Map values to 1 and 0, handling NaNs resulting from empty strings
                df[col] = df[col].map(lambda x: binary_mappings.get(x, x))
                print(f"✅ Columna '{col}' transformada a valores binarios")


    # Convert tenure column to numeric, coercing errors
    if 'tenure' in df.columns:
        df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
        print("✅ Columna 'tenure' convertida a numérica")

    # Convert TotalCharges column to numeric, coercing errors and filling NaNs
    if 'Total' in df.columns: # 'Total' after flattening 'Charges'
        df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
        df['Total'].fillna(df['Total'].median(), inplace=True) # Fill NaNs after coercion
        # Rename 'Total' to 'Cargo_Total' later in renaming step
        print("✅ Columna 'Total' convertida a numérica y NaNs rellenados")

    # Convert MonthlyCharges column to numeric, coercing errors
    if 'Monthly' in df.columns: # 'Monthly' after flattening 'Charges'
        df['Monthly'] = pd.to_numeric(df['Monthly'], errors='coerce')
        # Rename 'Monthly' to 'Cargo_Mensual' later in renaming step
        print("✅ Columna 'Monthly' convertida a numérica")


    # Renombrar columnas para mayor claridad (opcional)
    # Mapeo of common column names
    column_mapping = {
        'customerID': 'ID_Cliente',
        'gender': 'Genero',
        'SeniorCitizen': 'Ciudadano_Senior',
        'Partner': 'Tiene_Pareja',
        'Dependents': 'Tiene_Dependientes',
        'tenure': 'Antiguedad',
        'PhoneService': 'Servicio_Telefonico',
        'MultipleLines': 'Lineas_Multiples',
        'InternetService': 'Servicio_Internet',
        'OnlineSecurity': 'Seguridad_Online',
        'OnlineBackup': 'Respaldo_Online',
        'DeviceProtection': 'Proteccion_Dispositivo',
        'TechSupport': 'Soporte_Tecnico',
        'StreamingTV': 'Streaming_TV',
        'StreamingMovies': 'Streaming_Peliculas',
        'Contract': 'Tipo_Contrato',
        'PaperlessBilling': 'Facturacion_Sin_Papel',
        'PaymentMethod': 'Metodo_Pago',
        'Monthly': 'Cargo_Mensual', # Updated from MonthlyCharges
        'Total': 'Cargo_Total',     # Updated from TotalCharges
        'Churn': 'Evasion'
    }

    # Apply renaming only if the columns exist
    columns_to_rename = {k: v for k, v in column_mapping.items() if k in df.columns}
    if columns_to_rename:
        df = df.rename(columns=columns_to_rename)
        print("✅ Columnas renombradas para mayor claridad")
        print(f"Columnas renombradas: {list(columns_to_rename.keys())}")

    # Update column lists after transformation
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()


    # Check for and handle the empty string in 'Evasion' column
    if 'Evasion' in df.columns:
        if '' in df['Evasion'].unique():
            # Replace empty strings with NaN
            df['Evasion'].replace('', np.nan, inplace=True)
            # Drop rows with NaN in Evasion
            df.dropna(subset=['Evasion'], inplace=True)
            print("✅ Valores vacíos en la columna 'Evasion' tratados y filas correspondientes eliminadas.")
        # Convert 'Evasion' to numeric after handling empty strings
        # Ensure mapping is applied correctly
        if df['Evasion'].dtype == 'object':
             df['Evasion'] = df['Evasion'].map({'Yes': 1, 'No': 0})
             print("✅ Columna 'Evasion' convertida a valores binarios (1 para Yes, 0 para No).")


    # Verify the dataset after transformation
    print("\nDataset después de la transformación:")
    display(df.head())

    print(f"\nNuevo tamaño del dataset: {df.shape}")
    print(f"Columnas categóricas: {len(categorical_cols)}")
    print(f"Columnas numéricas: {len(numeric_cols)}")


# ===================================================================
# 3. CARGA Y ANÁLISIS (L - LOAD & ANALYSIS)
# ===================================================================

if data:
    print("\n" + "="*60)
    print("ETAPA 3: CARGA Y ANÁLISIS (LOAD & ANALYSIS)")
    print("="*60)

    # 3.1 Análisis descriptivo
    print("\n3.1 Análisis descriptivo:")

    # Estadísticas descriptivas para variables numéricas
    print("\nEstadísticas descriptivas - Variables numéricas:")
    display(df.describe())

    # Estadísticas descriptivas para variables categóricas
    print("\nEstadísticas descriptivas - Variables categóricas:")
    display(df.describe(include=['object']))

    # 3.2 Identificar la columna de churn/evasión
    churn_col = None
    churn_keywords = ['churn', 'evasi', 'baja', 'cancel']
    for col in df.columns:
        if any(keyword in col.lower() for keyword in churn_keywords):
            churn_col = col
            break

    if churn_col and churn_col in df.columns and df[churn_col].dtype == np.number:
        print(f"\n✅ Columna de churn/evasión identificada: '{churn_col}'")

        # 3.3 Análisis de la variable de churn
        print("\n3.3 Análisis de la variable de churn:")

        # Conteo de valores
        churn_counts = df[churn_col].value_counts()
        print("\nConteo de valores:")
        display(churn_counts)

        # Tasa de churn
        if 1 in churn_counts.index:
            churn_rate = churn_counts[1] / len(df) * 100
            print(f"\nTasa de churn: {churn_rate:.2f}%")

        # Visualización de la distribución de churn
        plt.figure(figsize=(10, 6))
        ax = sns.countplot(x=churn_col, data=df, palette='viridis')
        plt.title('Distribución de Evasión de Clientes', fontsize=16, fontweight='bold')
        plt.xlabel('Evasión', fontsize=12)
        plt.ylabel('Cantidad de Clientes', fontsize=12)

        # Add percentage labels
        total = len(df)
        for p in ax.patches:
            percentage = f'{100 * p.get_height() / total:.1f}%'
            ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom', fontsize=12, fontweight='bold')

        plt.tight_layout()
        plt.show()

        # 3.4 Análisis de churn por variables categóricas
        print("\n3.4 Análisis de churn por variables categóricas:")

        # Select important categorical columns (not more than 6 to avoid clutter)
        # Re-identify categorical columns after flattening
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
        important_cat_cols = [col for col in categorical_cols if col != churn_col][:6]

        for col in important_cat_cols:
            plt.figure(figsize=(12, 6))

            # Create contingency table
            contingency_table = pd.crosstab(df[col], df[churn_col], normalize='index') * 100

            # Stacked bar chart
            contingency_table.plot(kind='bar', stacked=True, colormap='viridis')
            plt.title(f'Tasa de Evasión por {col}', fontsize=14, fontweight='bold')
            plt.xlabel(col, fontsize=12)
            plt.ylabel('Porcentaje (%)', fontsize=12)
            plt.legend(title='Evasión', bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.xticks(rotation=45)

            # Add percentage labels
            for i, (index, row) in enumerate(contingency_table.iterrows()):
                cumulative = 0
                for j, value in enumerate(row):
                    if value > 5:  # Only show labels if the percentage is greater than 5%
                        plt.text(i, cumulative + value/2, f'{value:.1f}%',
                                ha='center', va='center', fontweight='bold', color='white')
                    cumulative += cumulative + value

            plt.tight_layout()
            plt.show()

            # Print churn rates by category
            churn_rate_by_category = df.groupby(col)[churn_col].mean().sort_values(ascending=False)
            print(f"\nTasa de evasión por {col} (ordenado de mayor a menor):")
            display(churn_rate_by_category)

        # 3.5 Análisis de churn por variables numéricas
        print("\n3.5 Análisis de churn por variables numéricas:")

        # Re-identify numeric columns after flattening and conversion
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
        important_num_cols = [col for col in numeric_cols if col != churn_col][:6]

        for col in important_num_cols:
            plt.figure(figsize=(15, 6))

            # Boxplot
            plt.subplot(1, 2, 1)
            sns.boxplot(x=churn_col, y=col, data=df, palette='viridis')
            plt.title(f'Boxplot de {col} por Evasión', fontsize=12, fontweight='bold')
            plt.xlabel('Evasión', fontsize=10)
            plt.ylabel(col, fontsize=10)

            # Histogram with KDE
            plt.subplot(1, 2, 2)
            for churn_value in df[churn_col].unique():
                subset = df[df[churn_col] == churn_value]
                sns.histplot(subset[col], kde=True, label=f'Evasión={churn_value}', alpha=0.6)
            plt.title(f'Distribución de {col} por Evasión', fontsize=12, fontweight='bold')
            plt.xlabel(col, fontsize=10)
            plt.ylabel('Frecuencia', fontsize=10)
            plt.legend()

            plt.tight_layout()
            plt.show()

            # Stats by churn group
            stats_by_churn = df.groupby(churn_col)[col].agg(['count', 'mean', 'std', 'min', 'max'])
            print(f"\nEstadísticas de {col} por estado de evasión:")
            display(stats_by_churn)

        # 3.6 Análisis de correlación
        print("\n3.6 Análisis de correlación:")

        # Select only numeric columns for correlation
        numeric_df = df.select_dtypes(include=['int64', 'float64'])

        if len(numeric_df.columns) > 1:
            # Calculate correlation matrix
            corr_matrix = numeric_df.corr()

            # Visualize correlation matrix
            plt.figure(figsize=(14, 10))
            sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f',
                       linewidths=0.5, annot_kws={"size": 10})
            plt.title('Matriz de Correlación', fontsize=16, fontweight='bold')
            plt.tight_layout()
            plt.show()

            # Correlation with the churn variable
            if churn_col in corr_matrix.columns:
                churn_corr = corr_matrix[churn_col].sort_values(ascending=False)
                print(f"\nCorrelación con {churn_col}:")
                display(churn_corr)

                # Visualize the most important correlations with churn
                plt.figure(figsize=(12, 6))
                churn_corr_without_self = churn_corr.drop(churn_col)
                colors = ['red' if x < 0 else 'green' for x in churn_corr_without_self]
                churn_corr_without_self.plot(kind='bar', color=colors)
                plt.title(f'Correlación de Variables con {churn_col}', fontsize=14, fontweight='bold')
                plt.ylabel('Coeficiente de Correlación', fontsize=12)
                plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
                plt.tight_layout()
                plt.show()
    else:
        print("\n⚠️ No se encontró una columna de churn/evasión o no es numérica en el dataset.")

    # 3.7 Guardar los datos procesados
    print("\n3.7 Guardar los datos procesados:")

    # Save to CSV format
    df.to_csv('TelecomX_Procesado.csv', index=False)
    print("✅ Datos guardados en formato CSV: 'TelecomX_Procesado.csv'")

    # Save to JSON format
    df.to_json('TelecomX_Procesado.json', orient='records')
    print("✅ Datos guardados en formato JSON: 'TelecomX_Procesado.json'")

    # Save to Excel format (requires openpyxl)
    try:
        df.to_excel('TelecomX_Procesado.xlsx', index=False)
        print("✅ Datos guardados en formato Excel: 'TelecomX_Procesado.xlsx'")
    except:
        print("⚠️ No se pudo guardar en formato Excel. Instala openpyxl: !pip install openpyxl")

# ===================================================================
# 4. INFORME FINAL
# ===================================================================

if data:
    print("\n" + "="*60)
    print("INFORME FINAL DEL PROCESO ETL")
    print("="*60)

    print("\n1. INTRODUCCIÓN:")
    print("El presente informe describe el proceso de ETL (Extract, Transform, Load) aplicado a los datos de Telecom X. "
          "El objetivo principal fue procesar los datos para permitir análisis más profundos y visualizaciones que ayuden a "
          "comprender los factores que influyen en la evasión de clientes (churn).")

    print("\n2. EXTRACCIÓN (EXTRACT):")
    print(f"- Se extrajeron datos desde la API de Telecom X disponible en: {url}")
    print(f"- Se obtuvo un dataset inicial con {df.shape[0]} filas y {df.shape[1]} columnas.")
    print("- Los datos se cargaron exitosamente en un DataFrame de pandas para su procesamiento.")

    print("\n3. TRANSFORMACIÓN (TRANSFORM):")
    print("- Se realizó una exploración inicial del dataset para entender su estructura.")
    print("- Se identificaron y trataron valores nulos utilizando estrategias apropiadas para cada tipo de variable.")
    if duplicate_rows > 0:
        print(f"- Se eliminaron {duplicate_rows} filas duplicadas.")
    if 'Cuentas_Diarias' in df.columns:
        print("- Se creó la columna 'Cuentas_Diarias' a partir de la facturación mensual.")
    print("- Se transformaron variables categóricas a valores binarios cuando fue necesario.")
    if columns_to_rename:
        print("- Se renombraron columnas para mayor claridad y consistencia.")

    print("\n4. CARGA Y ANÁLISIS (LOAD & ANALYSIS):")
    print("- Se realizó un análisis descriptivo completo de los datos.")
    if churn_col and churn_col in df.columns and df[churn_col].dtype == np.number:
        # Ensure churn_col is numeric before calculating mean
        if df[churn_col].dtype != np.number:
             # Assuming 'Yes'/'No' mapping happened correctly earlier
             df[churn_col] = df[churn_col].map({'Yes': 1, 'No': 0})
             df.dropna(subset=[churn_col], inplace=True) # Drop rows where mapping failed (e.g., empty strings)

        if churn_col in df.columns and df[churn_col].dtype == np.number:
            churn_rate = df[churn_col].mean() * 100
            print(f"- Se identificó una tasa de churn del {churn_rate:.2f}%.")
            print("- Se analizaron las variables categóricas y numéricas en relación con el churn.")
            print("- Se realizó un análisis de correlación para identificar las variables más relacionadas con el churn.")
        else:
             print("- La columna de churn/evasión no es numérica después de la transformación, no se puede calcular la tasa de churn.")
    else:
        print("- No se identificó una columna de churn en el dataset.")


    print("\n5. HALLAZGOS PRINCIPALES:")
    if churn_col and churn_col in df.columns and df[churn_col].dtype == np.number:
        # Identify variables with the biggest impact on churn
        # Re-identify important categorical columns after flattening
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
        important_cat_cols = [col for col in categorical_cols if col != churn_col][:6]

        if important_cat_cols:
            print("- Variables categóricas con mayor impacto en la evasión:")
            for col in important_cat_cols[:3]:
                if col in df.columns:
                    churn_rate_cat = df.groupby(col)[churn_col].mean().sort_values(ascending=False)
                    if not churn_rate_cat.empty:
                        max_category = churn_rate_cat.index[0]
                        max_rate = churn_rate_cat.iloc[0] * 100
                        print(f"  * {col}: La categoría '{max_category}' tiene la mayor tasa de evasión ({max_rate:.2f}%)")

        # Re-identify important numeric columns after flattening and conversion
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
        important_num_cols = [col for col in numeric_cols if col != churn_col][:6]

        if important_num_cols:
            print("- Variables numéricas con mayor impacto en la evasión:")
            for col in important_num_cols[:3]:
                if col in df.columns:
                    stats = df.groupby(churn_col)[col].mean()
                    if len(stats) == 2:
                        mean_churn = stats.iloc[1]
                        mean_no_churn = stats.iloc[0]
                        diff_pct = (mean_churn - mean_no_churn) / mean_no_churn * 100
                        print(f"  * {col}: Los clientes que abandonaron tienen un valor promedio {diff_pct:.2f}% {'mayor' if diff_pct > 0 else 'menor'} que los que permanecen")

    else:
        print("- No se pudo realizar el análisis de hallazgos principales porque la columna de churn no es numérica.")


    print("\n6. RECOMENDACIONES:")
    print("- Utilizar los datos procesados para desarrollar modelos predictivos de churn.")
    print("- Implementar estrategias de retención basadas en los factores identificados como relacionados con el churn.")
    print("- Realizar un seguimiento continuo de las métricas de churn para evaluar la efectividad de las estrategias implementadas.")
    print("- Considerar la recopilación de datos adicionales que puedan mejorar la precisión de los modelos predictivos.")
    print("- Automatizar el proceso de ETL para actualizaciones periódicas de los datos.")

    print("\n7. CONCLUSIONES:")
    print("- El proceso de ETL se completó exitosamente, transformando los datos crudos en un formato adecuado para el análisis.")
    if churn_col and churn_col in df.columns and df[churn_col].dtype == np.number:
        print("- Se identificaron variables con mayor relación con la evasión de clientes.")
        print("- Los análisis realizados proporcionan una base sólida para desarrollar estrategias de retención de clientes.")
    else:
         print("- No se pudo completar la conclusión sobre la relación entre variables y churn porque la columna de churn no es numérica.")
    print("- Los datos procesados están listos para análisis más avanzados y modelado predictivo.")


    print("\n" + "="*60)
    print("PROCESO ETL COMPLETADO EXITOSAMENTE")
    print("="*60)
```
